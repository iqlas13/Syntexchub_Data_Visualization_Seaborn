###Loading the Dataset

Once the dataset file is confirmed to be present in the working directory, the next step is to load the CSV file into a Pandas DataFrame. This allows us to view, analyze, and visualize the shopping behaviour data efficiently using Python libraries.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("Shopping_Behaviour_Dataset.csv")
df.head()
```

###Understanding the Dataset Structure

Before performing any data visualization, it is important to understand the structure of the dataset. This includes examining the number of rows and columns, data types of each feature, and identifying whether any missing values are present. This step helps in planning appropriate visualizations and analysis techniques.

```python
df.info()
```

###Dataset Summary

The dataset consists of 3,900 customer records with 18 features related to shopping behaviour. It includes a combination of numerical variables such as age, purchase amount, review rating, and previous purchases, along with categorical variables like gender, product category, season, payment method, and subscription status. There are no missing values, making the dataset well-suited for data visualization and exploratory analysis without additional cleaning.

```python
df.describe()
```

### Line Plot in Seaborn

A line plot is used to visualize trends or changes in data over a continuous variable. In this project, a line plot is used to observe how the purchase amount varies with age. Line plots are especially useful for identifying patterns and trends in numerical data.


```python
plt.figure(figsize=(8,5))
sns.lineplot(
    x="Age",
    y="Purchase Amount (USD)",
    data=df,
    color="teal"
)
plt.title("Purchase Amount vs Age")

plt.savefig("lineplot_purchase_amount_vs_age.png", dpi=300, bbox_inches="tight")
plt.show()
```

### Bar Plot in Seaborn

A bar plot is used to compare values across different categories. In this project, a bar plot is used to visualize the average purchase amount across different product categories. Bar plots are useful for identifying which categories contribute more to overall spending.


```python
plt.figure(figsize=(10,5))
sns.barplot(
    x="Category",
    y="Purchase Amount (USD)",
    hue="Category",
    data=df,
    palette="Set2",
    legend=False
)
plt.title("Average Purchase Amount by Category")
plt.xticks(rotation=45)

plt.savefig("barplot_avg_purchase_by_category.png", dpi=300, bbox_inches="tight")
plt.show()
```

### Histogram Plot in Seaborn

A histogram is used to understand the distribution of a numerical variable. In this project, a histogram is used to visualize the distribution of purchase amounts among customers. It helps identify the frequency, spread, and skewness of purchase values.


```python
plt.figure(figsize=(8,5))
sns.histplot(
    df["Purchase Amount (USD)"],
    bins=30,
    kde=True,
    color="steelblue"
)
plt.title("Distribution of Purchase Amount")
plt.xlabel("Purchase Amount (USD)")
plt.ylabel("Frequency")

plt.savefig("histogram_purchase_amount_distribution.png", dpi=300, bbox_inches="tight")
plt.show()
```

### Scatter Plot in Seaborn

A scatter plot is used to examine the relationship between two numerical variables. In this project, a scatter plot is used to analyze the relationship between customer age and purchase amount. Scatter plots help in identifying patterns, correlations, and outliers in the data.

```python
plt.figure(figsize=(8,5))
sns.scatterplot(
    x="Age",
    y="Purchase Amount (USD)",
    data=df,
    color="darkorange",
    alpha=0.6
)
plt.title("Age vs Purchase Amount")

plt.savefig("scatter_age_vs_purchase_amount.png", dpi=300, bbox_inches="tight")
plt.show()
```

### Heatmap Plot in Seaborn

A heatmap is used to visualize relationships between numerical variables using color intensity. In this project, a heatmap is created using the correlation matrix to understand how numerical features such as age, purchase amount, review rating, and previous purchases are related to each other.


```python
plt.figure(figsize=(8,5))

numeric_df = df.select_dtypes(include=["int64", "float64"])

sns.heatmap(
    numeric_df.corr(),
    annot=True,
    cmap="coolwarm",
    fmt=".2f"
)
plt.title("Correlation Heatmap")

plt.savefig("heatmap_numeric_correlations.png", dpi=300, bbox_inches="tight")
plt.show()
```

### Count Plot in Seaborn

A count plot is used to display the frequency of categorical variables. In this project, a count plot is used to visualize the number of customers across different genders. Count plots are helpful for understanding category-wise distribution in the dataset.

```python
plt.figure(figsize=(8,5))
sns.countplot(
    x="Gender",
    hue="Gender",
    data=df,
    palette="pastel",
    legend=False
)
plt.title("Customer Distribution by Gender")

plt.savefig("countplot_gender_distribution.png", dpi=300, bbox_inches="tight")
plt.show()
```

### Violin Plot in Seaborn

A violin plot combines features of a box plot and a density plot to show the distribution of data across categories. In this project, a violin plot is used to visualize the distribution of purchase amounts across different seasons, highlighting both data spread and density.


```python
plt.figure(figsize=(8,5))
sns.violinplot(
    x="Season",
    y="Purchase Amount (USD)",
    data=df,
    palette="muted"
)
plt.title("Purchase Amount Distribution Across Seasons")

plt.savefig("violinplot_purchase_by_season.png", dpi=300, bbox_inches="tight")
plt.show()
```

### Pair Plot in Seaborn

A pair plot is used to visualize pairwise relationships between numerical variables in a dataset. It helps in understanding correlations, trends, and distributions simultaneously. In this project, a pair plot is created using numerical features to explore relationships between age, purchase amount, review rating, and previous purchases.


```python
pair_plot = sns.pairplot(
    df[["Age", "Purchase Amount (USD)", "Review Rating", "Previous Purchases"]],
    diag_kind="kde"
)

pair_plot.savefig("pairplot_numerical_features.png", dpi=300)
plt.show()
```

### Strip Plot in Seaborn

A strip plot is used to visualize the distribution of individual data points across categories. In this project, a strip plot is used to observe how purchase amounts are distributed across different product categories. It helps in identifying patterns and overlaps in categorical data.


```python
plt.figure(figsize=(10,5))
sns.stripplot(
    x="Category",
    y="Purchase Amount (USD)",
    data=df,
    jitter=True,
    color="purple",
    alpha=0.6
)
plt.title("Purchase Amount Distribution by Category (Strip Plot)")
plt.xticks(rotation=45)

plt.savefig("stripplot_purchase_by_category.png", dpi=300, bbox_inches="tight")
plt.show()
```

### Box Plot in Seaborn

A box plot is used to summarize the distribution of numerical data using quartiles, median, and outliers. In this project, a box plot is used to compare purchase amounts across different product categories and to identify any outliers present in the data.


```python
plt.figure(figsize=(10,5))
sns.boxplot(
    x="Category",
    y="Purchase Amount (USD)",
    data=df,
    palette="Set3"
)
plt.title("Purchase Amount Distribution by Category (Box Plot)")
plt.xticks(rotation=45)

plt.savefig("boxplot_purchase_by_category.png", dpi=300, bbox_inches="tight")
plt.show()
```

### Factor Plot in Seaborn

A factor plot is used to visualize categorical data across different categories using statistical estimates. In this project, a factor plot is used to compare the average purchase amount across different seasons based on customer subscription status.

```python
factor_plot = sns.catplot(
    x="Season",
    y="Purchase Amount (USD)",
    hue="Subscription Status",
    data=df,
    kind="bar",
    palette="Set2",
    height=5,
    aspect=1.4
)

factor_plot.fig.suptitle(
    "Average Purchase Amount by Season and Subscription Status",
    y=1.02
)

factor_plot.savefig(
    "factorplot_purchase_by_season_subscription.png",
    dpi=300
)

plt.show()
```

### Cat Plot in Seaborn

A cat plot is a versatile categorical plotting function in Seaborn that allows visualization of relationships between categorical and numerical variables. In this project, a cat plot is used to compare purchase amount distributions across different seasons for each gender.

```python
cat_plot = sns.catplot(
    x="Season",
    y="Purchase Amount (USD)",
    hue="Gender",
    data=df,
    kind="box",
    palette="pastel",
    height=5,
    aspect=1.4
)

cat_plot.fig.suptitle(
    "Purchase Amount by Season and Gender",
    y=1.02
)

cat_plot.savefig(
    "catplot_purchase_by_season_gender.png",
    dpi=300
)

plt.show()
```

### Styling Plot in Seaborn

Styling in Seaborn allows customization of plots to improve readability and visual appeal. By using different themes, color palettes, and styles, visualizations can be made more professional and suitable for reports and presentations. In this project, Seaborn styling is applied to enhance the appearance of the plots.

```python
# Apply seaborn style
sns.set_style("whitegrid")
sns.set_palette("Set2")

plt.figure(figsize=(8,5))
sns.barplot(
    x="Season",
    y="Purchase Amount (USD)",
    hue="Season",
    data=df,
    legend=False
)

plt.title("Styled Bar Plot: Purchase Amount by Season")

plt.savefig("styled_barplot_purchase_by_season.png", dpi=300, bbox_inches="tight")
plt.show()
```

### Multiple Plot (Facet Grid) in Seaborn

A Facet Grid is used to create multiple subplots based on the values of one or more categorical variables. It allows comparison of data patterns across different subsets of the dataset. In this project, a Facet Grid is used to visualize purchase amount distribution across seasons, separated by gender.


```python
facet_grid = sns.FacetGrid(
    df,
    col="Season",
    row="Gender",
    height=3,
    aspect=1.2
)

facet_grid.map(
    sns.histplot,
    "Purchase Amount (USD)",
    bins=20,
    color="teal"
)

facet_grid.fig.suptitle(
    "Facet Grid: Purchase Amount Distribution by Season and Gender",
    y=1.05
)

facet_grid.savefig(
    "facetgrid_purchase_by_season_gender.png",
    dpi=300
)

plt.show()
```

The distribution of purchase amount shows a slightly right-skewed pattern, indicating that most customers make purchases within a lower to moderate price range, while a smaller number of customers spend significantly higher amounts. This skewness is clearly visible in the histogram and KDE plot, where the tail extends towards higher purchase values. Box plots and violin plots reveal the presence of outliers, representing customers with unusually high purchase amounts compared to the majority. The spread of purchase amounts varies across categories and seasons, with some categories showing wider variability than others. Overall, the analysis highlights unequal spending behavior among customers, with a few high-value purchases influencing the overall distribution.
